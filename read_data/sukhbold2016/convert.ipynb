```python
'''
This script replaces all '---' placeholders 
in the original Sukhbold data to '0.0' to
make the tables readable by read_sukhbold.f

Also "H3", "stability" and "NETWORK" entries are removed

-pkarpov
'''

import numpy as np
import sys
import os

path = '/home/pkarpov/scratch/Sukhbold2016/progenitor_models/'
to_delete = ["stability", "NETWORK", "He3"]

files = np.sort(os.listdir(path))

#path = './'
#files = ['test']
for progenitor in files:
    print(progenitor)
    file = path+progenitor
    header = []
    with open(file, 'r') as f:
        for header_line, line in enumerate(f):
            if header_line == 2: break
            header.append(line)
    
    h = header[-1].split("   ")
    while "" in h:
        h.remove("")

    for i in range(len(h)):
        h[i] = h[i].strip()                        
            
    allcontent = np.genfromtxt(file, skip_header=2, dtype=str)#, encoding=None)
    new_content = []        
    
    for content in allcontent:                
        content[0] = content[0][:-1]    

        for j in range(len(content)):
            if '---' in content[j]: content[j]='0.0'

        new_content.append(content)

    for d in to_delete:
        del_ind = h.index(d)
        if d == "He3":
            he4ind = h.index('He4')
            he3 = new_content[:,del_ind].astype(float)
            he4 = new_content[:,he4ind].astype(float)                                    
            hecomb = he3+he4
            new_content[:,he4ind] = ["%.16e"%val for val in hecomb]
        
        new_content = np.delete(new_content, del_ind, axis=1)  
        
        del h[del_ind]   
    header[-1] = '\t'.join(h) 
    
    new_content = new_content.astype(object)
    new_content[:,0] = new_content[:,0].astype(int)
    
    #progenitor = 'testout'
    with open(progenitor, 'w') as fout:
        for h in header: fout.write(h)
        fout.write('\n')
        np.savetxt(fout, new_content, delimiter='\t', fmt='%s')
```
