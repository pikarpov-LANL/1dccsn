```python
import os
import sys
import shutil
from subprocess import Popen, PIPE

class copy_rebuild:
    def __init__(self,m):
        self.src      = '/home/pkarpov/COLLAPSO1D'
        self.dst      = '/home/pkarpov/runs'
        self.suffix   = '_g9k_c8.4k_p_0.3k'        
        self.m        = m        
        self.run_path = f'{self.dst}/s{self.m}{self.suffix}'
        
        print(f'{self.run_path}')
        
    def copy(self,file):    
        shutil.copyfile(f'{self.src}/{file}', f'{self.dst}/s{self.m}{self.suffix}/{file}')
        
    def rebuild(self):
        # run 'make eos' and 'make project'
        os.chdir(f'{self.run_path}')
        
        p = Popen('make clean', shell=True, stdin=PIPE, stdout=PIPE, stderr=PIPE)
        output = p.stdout.read()
        p.stdout.close()
        
        p = Popen('make eos', shell=True, stdin=PIPE, stdout=PIPE, stderr=PIPE)
        output = p.stdout.read()        
        p.stdout.close()
        
        p = Popen('make project', shell=True, stdin=PIPE, stdout=PIPE, stderr=PIPE)
        output = p.stdout.read()
        p.stdout.close() 
        
        p = Popen('make readout', shell=True, stdin=PIPE, stdout=PIPE, stderr=PIPE)
        output = p.stdout.read()
        p.stdout.close()     

# Add custom hdf5 to path
#os.environ['LD_LIBRARY_PATH'] = os.environ['LD_LIBRARY_PATH']+':/home/pkarpov/Downloads/hdf5-1.12.2/hdf5/lib'

masses = [
          12.0,14.0,15.0,
          16.0,17.0,18.0]

for m in masses:            
    cr = copy_rebuild(m)
    cr.copy('Makefile')
    cr.copy('project/CMakeLists.txt')
    cr.copy('project/1dmlmix/CMakeLists.txt')
    cr.copy('project/1dmlmix/1dmlmix.f90')
    cr.copy('src/f90_bindings/CMakeLists.txt')   
    
    cr.rebuild()
```
